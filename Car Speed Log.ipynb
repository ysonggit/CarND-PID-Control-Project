```python
import matplotlib.pyplot as plt
import csv

def plotSteeringSpeedCte(fname, start=0, end=100000):
    c = []
    y = []
    x = []
    s = []

    with open(fname,'r') as csvfile:
        plots = csv.reader(csvfile, delimiter=',')
        i = 0
        for row in plots:
            i+= 1
            x.append(i)
            c.append(float(row[0])) # cte
            y.append(float(row[1])) # steering
            s.append(float(row[2])) # speed

    end_idx = min(end, len(x))
    f, axarr = plt.subplots(2, sharex=True, figsize=(20, 10))
    f.suptitle('Car Steering Angle/CTE and Speed')
    axarr[0].plot(x[start:end_idx],y[start:end_idx], label='steering angle')
    axarr[0].plot(x[start:end_idx],c[start:end_idx], label="CTE")   
    axarr[0].legend(['steering angle', 'CTE'], loc='upper left')
    axarr[1].plot(x[start:end_idx],s[start:end_idx], label='speed')
    axarr[1].legend(['speed'], loc='upper left')
    
plotSteeringSpeedCte('logs/car_param.log.20-01-2019-13-05-44')
```

```python
def plotSteeringCte(fname, start=0, end=100000):
    c = []
    y = []
    x = []

    with open(fname,'r') as csvfile:
        plots = csv.reader(csvfile, delimiter=',')
        i = 0
        for row in plots:
            i+= 1
            x.append(i)
            c.append(float(row[0])) # cte
            y.append(float(row[1])) # steering angle

    end_idx = min(end, len(x))
    plt.plot(x[start:end_idx],y[start:end_idx], label='steering angle')
    plt.plot(x[start:end_idx],c[start:end_idx], label="CTE")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Car Params')
    plt.legend(['steering angle', 'CTE'], loc='upper left')
    plt.show()
    
plotSteeringCte('build/car_param.log', 3000,4500)
```
